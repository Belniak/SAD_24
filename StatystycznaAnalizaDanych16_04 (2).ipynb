```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import folium
import json
```

```python
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/SAD/POSE-205.xlsx')
df = df.dropna()
df1 = pd.read_json(r'/content/drive/MyDrive/Colab Notebooks/SAD/R.json')
```

# Zmiana zanieczyszczenia w czasie na przykładzie jednej wybranej placówki.

```python
df_miasto = df[df['name'] == 'SZKOŁA PODSTAWOWA NR 36 IM. CZESŁAWA MIŁOSZA W RYBNIKU']
```

```python
sns.lineplot(data=df_miasto,x='date', y='pm10_max', label='Maksymalne zanieczyszczenie')
sns.lineplot(data=df_miasto,x='date', y='pm10_min', label='Minimalne zanieczyszczenie')
plt.xlabel('Czas')
plt.ylabel('Wartość')
plt.xticks(df_miasto['date'][::50], rotation=60)
plt.legend()
plt.show()
```

# Lokalizacje placówek na mapie.

```python
lokalizacje = df1['geolokalizacja']
punkty = []
for lokalizacja in lokalizacje:
  punkty.append([lokalizacja['latitude'], lokalizacja['longitude']])
```

```python
lokalizacje = df1['geolokalizacja']
punkty = []
for lokalizacja in lokalizacje:
  punkty.append([lokalizacja['latitude'], lokalizacja['longitude']])

mymap = folium.Map(location=[52.2297, 21.0122], zoom_start=6)

for punkt in punkty:
  folium.Marker(location=punkt).add_to(mymap)

mymap
```
